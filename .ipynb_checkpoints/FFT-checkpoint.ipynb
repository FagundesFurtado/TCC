```python
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import numpy as np
import pylab
import pyaudio
import wave
import struct
import time
from scipy.io import wavfile
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python

def data_to_float(astr, size, channels):
    a = struct.unpack("%ih" % (size* channels), astr)
    a = [float(val) / pow(2, 15) for val in a]
    return a


def wholeSong(path, play=False):
    p = pyaudio.PyAudio()
    wf = wave.open(path, 'rb')
    data = []
    raw_data = []
    chunk = 1024
    time = 1
    rateDefault = wf.getframerate()
    chunk = int(rateDefault * time)
    channels = wf.getnchannels()
    buffer = chunk
    stream = p.open(format =
                    p.get_format_from_width(wf.getsampwidth()),
                    channels = wf.getnchannels(),
                    rate = wf.getframerate(),
                    output = True)
    dataStream = wf.readframes(buffer)
    while dataStream != '':
        if(len(dataStream)== 0):
            break
        stream.write(dataStream)
        print("Size", len(dataStream))
        metade = len(dataStream)//(channels *2)
        print("metade ", metade)
        data.append(data_to_float(dataStream, metade,channels))
        raw_data.append(dataStream)
        dataStream = wf.readframes(buffer)
    stream.close()    
    return rateDefault, data, raw_data, wf
def playData(data, wf):
    p = pyaudio.PyAudio()
    stream = p.open(format =
                    p.get_format_from_width(wf.getsampwidth()),
                    channels = wf.getnchannels(),
                    rate = wf.getframerate(),
                    output = True)
    stream.write(data)
    stream.close()
def plotSignalAndPeakFrequency(rate, data, play, raw,  wf):
    f, P = signal.periodogram(data, rate, axis=0, detrend='linear') # Frequencies and PSD

    fig, axe = plt.subplots()
    axe.semilogy(f, P)
    axe.set_xlim([0,500])
    max = np.max(data)
    axe.set_ylim([1e-10 , 2*max])
    #axe.set_ylim([1e-9, 1e10])
    axe.set_xlabel(r'Frequency, $\nu$ $[\mathrm{Hz}]$')
    axe.set_ylabel(r'PSD, $P$ $[\mathrm{AU^2Hz}^{-1}]$')
    axe.set_title('Periodogram')
    axe.grid(which='both')
    position = np.where(P == np.amax(P))
    print("Frequencia de Pico ", f[position[0][0]])
    if(play):
        playData(raw,wf)
    
    return f[position[0][0]]
```

```python
rate, arquivo, raw_data, wf  = wholeSong('440t.wav')
```

```python
frequencias = []

for i in range(len(arquivo)):
    posicao = i
    freq = plotSignalAndPeakFrequency(rate, arquivo[posicao], True, raw_data[posicao],wf)
    frequencias.append(freq)
    time.sleep(1)
```
