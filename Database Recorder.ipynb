```python
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import numpy as np
import pylab
import pyaudio
import wave
import struct
import time
from scipy.io import wavfile
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def data_to_float(astr, size, channels):
    a = struct.unpack("%ih" % (size* channels), astr)
    a = [float(val) / pow(2, 15) for val in a]
    return a
def playAudio(data):
    p = pyaudio.PyAudio()
    FORMAT = pyaudio.paInt16
    CHANNELS = 1
    RATE = 44100
    stream = p.open(format =
                    FORMAT,
                    channels = CHANNELS,
                    rate = RATE,
                    output = True)
    print("Playing")
    for i in data:
        stream.write(i)
    print("Played")
    stream.close()
def gravarNotas(RECORD_SECONDS=5):
    chunk = 1024
    FORMAT = pyaudio.paInt16
    CHANNELS = 1
    RATE = 44100
    WAVE_OUTPUT_FILENAME = "output.wav"

    p = pyaudio.PyAudio()

    stream = p.open(format = FORMAT,
                    channels = CHANNELS,
                    rate = RATE,
                    input = True,
                    output = True,
                    frames_per_buffer = chunk)

    all = []
    converted = []
    for i in range(0, RATE // chunk * RECORD_SECONDS):
        data = stream.read(chunk)
        all.append(data)
        converted.append(data_to_float(data, len(data)//2, 1))
    stream.close()
    p.terminate()
    return all, converted
def save_wav(file_name, audio):
    wav_file=wave.open(file_name,"w")

    nchannels = 1

    sampwidth = 2
    sample_rate = 44100
    nframes = len(audio)
    comptype = "NONE"
    compname = "not compressed"
    wav_file.setparams((nchannels, sampwidth, sample_rate, nframes, comptype, compname))

    for sample in audio:
        for i in sample:
            wav_file.writeframes(struct.pack('h', int( i * 32767 )))
    wav_file.close()
    print("Salvou arquivo")
    return
```
